```python
import numpy as np
import pandas as pd
import seaborn as sns # Import Seaborn

df = pd.read_csv('./visualization/assignment/data/airbnb_hw.csv')

var = df['Price'] # Categorical variable example
var = var.str.replace(',','') #removing commas
var = pd.to_numeric(var, errors='coerce') #coerce the variable to numeric
# Create a missing dummy:
df['Price'] = var
del var
```

```python
print(df.shape, '\n') # returns (rows,columns)
print(df.head, '\n')
```

```python
print(pd.crosstab(df['Property Type'], df['Room Type']))
```

```python
sns.histplot(df['Price'])
```

```python
sns.kdeplot(df['Price'])
```

```python
sns.boxplot(data = df, x = 'Price')
```

```python
df['Price'].describe()

df['log_pr'] = np.log(df['Price'])
var1 = 'log_pr'
```

```python
sns.histplot(df[var1])
```

```python
sns.kdeplot(df[var1])
```

```python
sns.boxplot(data = df, x = var1)
```

```python
df[var1].describe()

sns.scatterplot(data = df, y= var1,x= 'Beds')
```

```python
df.loc[:,['Beds', 'Price']].groupby('Beds').describe()

sns.scatterplot(data = df, x= 'Beds',y= var1, hue = 'Room Type')
```

```python
df.loc[:,['Room Type', 'Price']].groupby('Room Type').describe()

sns.jointplot(data=df,x='Beds',y=var1,kind='hex')
```
