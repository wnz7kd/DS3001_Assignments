````python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans # Import kmc
import matplotlib.pyplot as plt

def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

````python
sns.scatterplot(data = df0_125, x = 'x1',y='x2',hue='group',style='group')
plt.title('Dataset with Noise Level 0.125')
plt.show()
````

````python
sns.scatterplot(data = df0_25, x = 'x1',y='x2',hue='group',style='group')
plt.title('Dataset with Noise Level 0.25')
plt.show()
````

````python
sns.scatterplot(data = df0_5, x = 'x1',y='x2',hue='group',style='group')
plt.title('Dataset with Noise Level 0.5')
plt.show()
````

````python
sns.scatterplot(data = df1_0, x = 'x1',y='x2',hue='group',style='group')
plt.title('Dataset with Noise Level 1')
plt.show()
````

````python
sns.scatterplot(data = df2_0, x = 'x1',y='x2',hue='group',style='group')
plt.title('Dataset with Noise Level 2')
plt.show()
````

At higher noise levels, the points from different clusters start to mix, leading to less clear boundaries between clusters in the scatterplots. In other words, more variability in the data makes it harder for k-means clustering to accurately separate the clusters.

````python
def maxmin(x): # Max-min scaler function
    x = (x-min(x))/(max(x)-min(x))
    return x
def scree(data):
  X = data.loc[ : , ['x1','x2'] ] # Select data to cluster
  X = X.apply(maxmin) # Normalize the data
  #scree plot:
  k_bar = 10
  k_grid = np.arange(1,k_bar+1) # range of k's to consider
  SSE = np.zeros(k_bar) # training error
  for k in range(k_bar):
      model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0) # Create a model for k
      model = model.fit(X) # fitting
      SSE[k] = model.inertia_ # SSE
  scree_plot, axes = plt.subplots()
  sns.lineplot(x=k_grid, y=SSE).set_title('Scree Plot')
  axes.set_ylim(0, 35)
````

````python
scree(data=df0_125 )
````

There is a clear elbow at k = 3 here, where after that very little improvement is made in terms of dropping the SSE. The marginal benefit of an additional cluster is close to zero.

````python
scree(data=df0_25)
````

````python
scree(data=df0_5)
````

````python
scree(data=df1_0)
````

````python
scree(data=df2_0)
````

By the last data set with the highest noise, the elbow is much less defined and even the first and second k's don't cause significant drops in SSE, like they did in the first data set. This makes it harder to choose a definitive k, as the scree visual is more ambiguous due to higher noise.

To decide on an optimal k, people look for a spot where the reduction in SSE of going from k-1 to k is significantly higher than going from k to k+1. The "elbow" in the scree plot represents this point of diminishing returns. The intuition is that beyond this elbow point, the increase in model complexity (more clusters) doesn't provide substantial benefits in terms of reducing SSE. Instead, it may lead to overfitting or capturing noise in the data. This data clearly shows that k* become much less obvious as variability increases. The data is more convoluded, making this ideal k harder to decipher.
